This code creates a function that takes several hyperparameters related to Word2Vec and returns a Word2Vec model. Additionally, it provides a set of hyperparameters in lists, that is next to be iterated over various combinations of these hyperparameters to tune the Word2Vec model.

```python
import numpy as np
from gensim.models import Word2Vec

# Define different hyperparameters to tune
embedding_sizes = [50, 100, 150]
window_sizes = [3, 5, 7]
min_word_counts = [1, 2, 3]
epochs_list = [5, 10, 15]

# Function to create word embeddings
def create_word_embeddings(embedding_size, window_size, min_word_count, epochs):
    combined_content = train_df['content'].append(val_df['content'], ignore_index=True)
    word2vec_model = Word2Vec(
        combined_content,
        vector_size=embedding_size,
        window=window_size,
        min_count=min_word_count,
        epochs=epochs
    )
    return word2vec_model
```

 The following code is a part of a hyperparameter tuning process for the previously created Word2Vec model. It creates embeddings, clusters the word vectors using KMeans, applies t-SNE for visualization, and displays the clusters in a 2D scatter plot for different combinations of hyperparameters.

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Loop to tune Word2Vec hyperparameters
for embedding_size in embedding_sizes:
    for window_size in window_sizes:
        for min_word_count in min_word_counts:
            for epochs in epochs_list:
                word2vec_model = create_word_embeddings(embedding_size, window_size, min_word_count, epochs)

                # Get word vectors and their corresponding words
                word_vectors = word2vec_model.wv.vectors[200:350]
                words = word2vec_model.wv.index_to_key[200:350]

                # Cluster the word vectors using KMeans
                num_clusters = 5  # Adjust the number of clusters as needed
                kmeans = KMeans(n_clusters=num_clusters, random_state=42)
                cluster_indices = kmeans.fit_predict(word_vectors)

                # Visualize clusters using t-SNE for dimensionality reduction to 2D
                tsne = TSNE(n_components=2, random_state=42)
                word_vec_2d = tsne.fit_transform(word_vectors)

                print(f"Parameters: embedding_size={embedding_size}, window_size={window_size}, min_word_count={min_word_count}, epochs={epochs}")
                # Scatter plot for visualizing clusters in 2D
                plt.figure(figsize=(18, 10))
                for i in range(num_clusters):
                    cluster_points = word_vec_2d[cluster_indices == i]
                    cluster_words = np.array(words)[cluster_indices == i]
                    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {i}")
                    for j, word in enumerate(cluster_words):
                        plt.text(cluster_points[j, 0], cluster_points[j, 1], word, fontsize=8)
                plt.legend()
                plt.title('Word Clusters Visualization (t-SNE)')
                plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from gensim.models import Word2Vec
from sklearn.metrics import accuracy_score

# Define different hyperparameters to tune
embedding_sizes = [50, 100, 150]
window_sizes = [3, 5, 7]
min_word_counts = [1, 2, 3]
epochs_list = [5, 10, 15]

best_accuracy = 0
best_params = {}
# Function to create word embeddings
def create_word_embeddings( embedding_size, window_size, min_word_count, epochs):
  combined_content = train_sentence_df['content'].append(val_sentence_df['content'], ignore_index=True)
  word2vec_model = Word2Vec(
      combined_content,
      vector_size=embedding_size,
      window=window_size,
      min_count=min_word_count,
      epochs=epochs
  )
  return word2vec_model

input_size = 100  # Assuming word embeddings of size 100
hidden_size = 128  # This can be adjusted based on experimentation
output_size = 3  # For a 3-class sentiment analysis (for example
# Loop to tune Word2Vec hyperparameters
for embedding_size in embedding_sizes:
    for window_size in window_sizes:
        for min_word_count in min_word_counts:
            for epochs in epochs_list:
                word2vec_model = create_word_embeddings( embedding_size, window_size, min_word_count, epochs)

                def generate_embeddings(data):
                    embeddings = []
                    for words_list in data:
                        valid_words = [word for word in words_list if word in word2vec_model.wv]
                        if valid_words:
                            # Only append if there are valid words
                            embeddings.append(np.mean([word2vec_model.wv[word] for word in valid_words], axis=0))
                        else:
                            # Handle the case where there are no valid words
                            embeddings.append(np.zeros(embedding_size, dtype=np.float32))
                    return np.array(embeddings)

                # Construct embeddings for X_train and X_val
                X_train = generate_embeddings(train_sentence_df['content'])
                X_val = generate_embeddings(val_sentence_df['content'])

                y_train = train_sentence_df['sentiment_label'].values
                y_val = val_sentence_df['sentiment_label'].values

                # Design neural network for sentiment classification
                class YourSentimentClassifier(nn.Module):
                    def __init__(self, input_size, hidden_size, output_size):
                      super(YourSentimentClassifier, self).__init__()
                      self.fc1 = nn.Linear(input_size, hidden_size)
                      self.relu = nn.ReLU()
                      self.fc2 = nn.Linear(hidden_size, output_size)

                    def forward(self, x):
                        out = self.fc1(x)
                        out = self.relu(out)
                        out = self.fc2(out)
                        return out

                # Initialize model, loss function, and optimizer
                model = YourSentimentClassifier(embedding_size, hidden_size, output_size)
                criterion = nn.CrossEntropyLoss()
                optimizer = optim.Adam(model.parameters(), lr=0.001)

                # Training the neural network
                num_epochs = 200  # Set the number of epochs for training
                for epoch in range(num_epochs):
                    optimizer.zero_grad()
                    outputs = model(torch.tensor(X_train))
                    loss = criterion(outputs, torch.tensor(y_train))
                    loss.backward()
                    optimizer.step()

                # Evaluation on validation set
                with torch.no_grad():
                    outputs = model(torch.tensor(X_val, dtype=torch.float32))
                    _, predicted = torch.max(outputs, 1)
                    accuracy = accuracy_score(y_val, predicted.numpy())

                # Check if this combination of hyperparameters gives a better accuracy
                if accuracy >= best_accuracy:
                    best_accuracy = accuracy
                    best_params = {
                        'embedding_size': embedding_size,
                        'window_size': window_size,
                        'min_word_count': min_word_count,
                        'epochs': epochs
                    }
                    print(best_params)
                    print(accuracy)

print("Best Accuracy:", best_accuracy)
print("Best Parameters:", best_params)
```

### 3.3 BERT

Firstly, we create a basic tokenizer class, which constructs a vocabulary based on unique words from the column 'content' of original dataset. Then, it uses the tokenizer to convert sentences into sequences of token IDs.

```python
class SimpleTokenizer:
    def __init__(self, vocab):
        self.vocab = vocab

    def tokenize(self, sentences):
        tokenized_sentences = []
        for sentence in sentences:
            # Simple tokenization: split by space and convert tokens to IDs
            tokens = sentence.lower().split()  # Basic tokenization (you might need more sophisticated methods)
            token_ids = [self.vocab.get(token, self.vocab['[UNK]']) for token in tokens]
            tokenized_sentences.append(token_ids)
        return tokenized_sentences

# Extract unique words from df_sentences['content'] to build the vocabulary
unique_words = set()
for sentences in df_sentences['content']:
    for sentence in sentences:
        unique_words.update(sentence.lower().split())

# Add special tokens and convert unique words to token IDs
vocab = {
    '[PAD]': 0,
    '[UNK]': 1,
    '[CLS]': 2,
    '[SEP]': 3,
}

# Assign IDs to unique words
for i, word in enumerate(unique_words, len(vocab)):
    vocab[word] = i

# Initialize your tokenizer
tokenizer = SimpleTokenizer(vocab)

# Tokenize sentences from df_sentences['content']
tokenized_sentences = [tokenizer.tokenize(sentence) for sentence in df_sentences['content']]
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ScratchBERT(nn.Module):
    def __init__(self, vocab_size, hidden_size=768, num_layers=12, num_heads=12):
        super(ScratchBERT, self).__init__()
        self.embedding = nn.Embedding(vocab_size, hidden_size)
        self.encoder_layers = nn.ModuleList([
            EncoderLayer(hidden_size, num_heads) for _ in range(num_layers)
        ])
        self.pad_token_id = 0  # Assuming 0 is the index for the padding token
        self.max_length = max_length  # Assuming 0 is the index for the padding token

    def forward(self, input_ids):
        # Truncate or pad sequences to max_length
        if input_ids.size(1) > self.max_length:
            input_ids = input_ids[:, :self.max_length]
        elif input_ids.size(1) < self.max_length:
            pad_amount = self.max_length - input_ids.size(1)
            padding = torch.full((input_ids.size(0), pad_amount), self.pad_token_id, dtype=torch.long)
            input_ids = torch.cat([input_ids, padding], dim=1)

        # Mask indicating positions of padded tokens
        attention_mask = (input_ids != self.pad_token_id).unsqueeze(1).unsqueeze(2)

        embedded = self.embedding(input_ids)
        encoded = embedded  # For simplicity, initial embeddings are passed to the encoder directly

        for layer in self.encoder_layers:
            encoded = layer(encoded, attention_mask)  # Applying each layer of the encoder

        return encoded

class EncoderLayer(nn.Module):
    def __init__(self, hidden_size, num_heads):
        super(EncoderLayer, self).__init__()
        self.self_attention = MultiHeadSelfAttention(hidden_size, num_heads)
        self.feedforward = FeedForward(hidden_size)

    def forward(self, x, attention_mask):
        attention_output = self.self_attention(x, attention_mask)
        feedforward_output = self.feedforward(attention_output)
        return feedforward_output

class MultiHeadSelfAttention(nn.Module):
    def __init__(self, hidden_size, num_heads):
        super(MultiHeadSelfAttention, self).__init__()
        self.hidden_size = hidden_size
        self.num_heads = num_heads

        # Linear projections for query, key, and value
        self.linear_query = nn.Linear(hidden_size, hidden_size)
        self.linear_key = nn.Linear(hidden_size, hidden_size)
        self.linear_value = nn.Linear(hidden_size, hidden_size)

        # Final linear transformation
        self.linear_final = nn.Linear(hidden_size, hidden_size)

    def forward(self, x):
        # Assume x has shape (batch_size, sequence_length, hidden_size)

        # Linearly project inputs for queries, keys, and values
        query = self.linear_query(x)
        key = self.linear_key(x)
        value = self.linear_value(x)

        # Reshape/query for multi-heads
        query = query.view(query.size(0), -1, self.num_heads, self.hidden_size // self.num_heads).transpose(1, 2)
        key = key.view(key.size(0), -1, self.num_heads, self.hidden_size // self.num_heads).transpose(1, 2)
        value = value.view(value.size(0), -1, self.num_heads, self.hidden_size // self.num_heads).transpose(1, 2)

        # Calculate scaled dot-product attention
        attention_scores = torch.matmul(query, key.transpose(-2, -1)) / (self.hidden_size ** 0.5)
        attention_weights = F.softmax(attention_scores, dim=-1)
        attention_output = torch.matmul(attention_weights, value)

        # Reshape and concatenate attention outputs
        attention_output = attention_output.transpose(1, 2).contiguous().view(x.size(0), -1, self.hidden_size)

        # Linear projection for final output
        output = self.linear_final(attention_output)
        return output

class FeedForward(nn.Module):
    def __init__(self, hidden_size):
        super(FeedForward, self).__init__()
        self.fc1 = nn.Linear(hidden_size, hidden_size * 4)  # Assuming hidden_size * 4 as intermediate dimension
        self.fc2 = nn.Linear(hidden_size * 4, hidden_size)  # Back to original hidden_size

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Instantiate your ScratchBERT model
# model = ScratchBERT(vocab_size=len(vocab))
```
